```python
def insideTrench(x, y):
    inx = txmin <= x <= txmax
    iny = tymin <= y <= tymax
    return inx and iny

def overshotTrench(x,y):
    outx = x > txmax
    outy = y < tymin
    return outx or outy
    
```

```python
def simulate(vx, vy):
    velx,vely = vx,vy
    x, y = 0,0
    steps = 0
    maxy = 0
    while not overshotTrench(x, y):
        x += velx
        y += vely
        velx += -1 if velx > 0 else 1
        vely -= 1
        maxy = max(maxy, y)
        steps += 1
        if insideTrench(x, y):
            return steps, maxy
    return False


```

```python
def simulateY(vy, vx= 19):
    velx,vely = vx,vy
    x, y = 0,0
    steps = 0
    maxy = 0
    intrench = False
    withinX = False
    while not overshotTrench(x, y):
        x += velx
        y += vely
        velx += -1 if velx > 0 else 1
        vely -= 1
        maxy = max(maxy, y)
        steps += 1
        withinX = withinX or txmin <= x <= txmax
        if not intrench and insideTrench(x, y):
            # print ("Vy", vy,"reached the trench in", steps, "steps and peaked at", maxy)
            intrench = True
    if intrench:
        # print ("Vy", vy,"overshot the trench after", steps, "steps.")
        return True, maxy
    elif withinX:
        return False, maxy
    else:
        return True, -1
```

```python
vy = 0
maxys = []
for vx in range(19, 202):
    while True:
        res = simulateY(vy, vx)
        if res[0]:
            maxys.append(res[1])
        else: break
        vy += 1
print(sorted(maxys, reverse=True))
```
