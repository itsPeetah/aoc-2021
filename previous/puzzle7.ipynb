```python
initialCrabPositions = sorted(list(map(int, open("puzzle7.txt").readline().split(','))))
crabsPerPosition = {}
for c in initialCrabPositions:
    if c in crabsPerPosition: crabsPerPosition[c] += 1
    else: crabsPerPosition[c] = 1
```

```python
minPos = min(initialCrabPositions)
maxPos = max(initialCrabPositions)
minFuel = sum(initialCrabPositions)
alignPos = -1
for x in range(minPos, maxPos + 1):
    curFuel = sum([abs(c - x) for c in initialCrabPositions])
    if curFuel < minFuel:
        print(f"Getting closer: {alignPos} -> {x} (fuel used:{minFuel}->{curFuel})")
        minFuel = curFuel
        alignPos = x
print(minFuel, alignPos)
```

```python
def fuel_usage(position, target):
    a = abs(position - target)
    n = a * (a + 1)
    return n / 2
```

```python
minPos = min(initialCrabPositions)
maxPos = max(initialCrabPositions)
minFuel = sum(initialCrabPositions) ** 2
alignPos = -1
for x in range(minPos, maxPos + 1):
    curFuel = sum([fuel_usage(c, x) for c in initialCrabPositions])
    if curFuel < minFuel:
        print(f"Getting closer: {alignPos} -> {x} (fuel used:{minFuel}->{curFuel})")
        minFuel = curFuel
        alignPos = x
print(minFuel, alignPos)
```
